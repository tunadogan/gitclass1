```python
numbers = [1,3,  7, 4, 6, 3, 0, 3, 6, 3, 7, 6, 6] 
max(numbers, key=numbers.count)                                # ilk olanı alır. numberin içinde tekrar sayıları key oluyor.
maximum = max(numbers, key=numbers.count)                         # max fonk u bir tane iterable elemanı döndürür.
cnt= numbers.count(maximum)
print (f"the most frequent number is {maximum} and it was {cnt} times repeated")
```
