```python
def duplicate(nums):     #aynı eleman var ise False çıktısı verecek fonksiyon tanımı
    num_set=set(nums)    # set() ile aynı elemanları birleştirebiliriz.
    return len (nums) != len(num_set)  #setlenmiş ile kullanıcıdan alınan karşılaştırılır. 
```

```python
print(duplicate([1,2,3,1]))
print(duplicate([1,2,3,4]))
print(duplicate([1,1,1,3,3,4,3,2,4,2]))
```

```python
def missing(find): # aranan numara için fonksiyon adı.
    k = len(find)  #girilen listenin eleman sayısı bulunur.
    tam = ((k + 1)*(k + 2))/2  # sıralı liste olduğu için başı sonu belli elemanların toplamı bulunur.
    son_sum = sum(find) # kullanıcıdan alınan liste elemanları toplamı bulunur.
    return list(tam - son_sum) # normalde olan toplam ile verilen liste toplamının farkı eksik elemanı verecek.
```

```python
missing([1,3,2,6])
```

```python
def singleNumber(nums):
    return nums.reduce(lambda x, y: x ^ y, nums, 0)
```

```python
singleNumber([4,1,2,1,2])
```

```python
def singleNumber(nums):
    ones = 0
    twos = 0

    for num in nums:
        ones ^= (num & ~twos)
        twos ^= (num & ~ones)

    return ones
```

```python
singleNumber([4,1,2,1,2])
```
